```python
#Fiftyone lib
import fiftyone as fo
import fiftyone.zoo as foz

#Mediapipe lib
from google.colab import files
import os
import json
import tensorflow as tf
assert tf.__version__.startswith('2')
from mediapipe_model_maker import object_detector

# The directory to which to write the exported dataset
export_dir_train = '/content/sample_data/coco-2017/train'
export_dir_validation = '/content/sample_data/coco-2017/validation'

# The name of the sample field containing the label that you wish to export
# Used when exporting labeled datasets (e.g., classification or detection)
label_field = "ground_truth"  # for example

# The type of dataset to export
# Any subclass of `fiftyone.types.Dataset` is supported
dataset_type = fo.types.COCODetectionDataset  # for example

#Samples to downlaod
max_samples_train = 1000
max_samples_validation = 100

#Download train dataset
dataset_train = foz.load_zoo_dataset(
    "voc-2012",
    split="train",
    shuffle=True,
    max_samples = max_samples_train,
)

#Export train dataset
dataset_train.export(
    export_dir=export_dir_train,
    dataset_type=dataset_type,
    label_field=label_field,
)

#rename "data" folder into "images"
%mv /content/sample_data/coco-2017/train/data /content/sample_data/coco-2017/train/images

#Load train dataset into mediapipe
train_data = object_detector.Dataset.from_coco_folder(export_dir_train, cache_dir="/tmp/od_data/train")
print("train_data size: ", train_data.size)

#Remove export
%rm -r /content/sample_data/coco-2017/train

#Download validation dataset
dataset_validation = foz.load_zoo_dataset(
    "voc-2012",
    split="validation",
    shuffle=True,
    max_samples = max_samples_validation,
)

#Export validation dataset
dataset_validation.export(
    export_dir=export_dir_validation,
    dataset_type=dataset_type,
    label_field=label_field,
)

#rename "data" folder into "images"
%mv /content/sample_data/coco-2017/validation/data /content/sample_data/coco-2017/validation/images

#Load validation dataset into mediapipe
validation_data = object_detector.Dataset.from_coco_folder(export_dir_validation, cache_dir="/tmp/od_data/validation")
print("validation_data size: ", validation_data.size)

%rm -r /content/sample_data/coco-2017/validation
```

```python
#Load pre trained model
spec = object_detector.SupportedModels.MOBILENET_MULTI_AVG
hparams = object_detector.HParams(export_dir='exported_model')
options = object_detector.ObjectDetectorOptions(
    supported_model=spec,
    hparams=hparams
)
```

```python
#Train model
model = object_detector.ObjectDetector.create(
    train_data=train_data,
    validation_data=validation_data,
    options=options)
```

```python
#Export model
model.export_model("model_original.tflite")
%mv exported_model/model_original.tflite /content/drive/MyDrive/sharing_storage/model_original_voc_2012.tflite
```

```python
#Quantization
qat_hparams = object_detector.QATHParams(learning_rate=0.3, batch_size=4, epochs=10, decay_steps=6, decay_rate=0.96)
model.quantization_aware_training(train_data, validation_data, qat_hparams=qat_hparams)
#Skip validation because takes a long time
#qat_loss, qat_coco_metrics = model.evaluate(validation_data)
#print(f"QAT validation loss: {qat_loss}")
#print(f"QAT validation coco metrics: {qat_coco_metrics}")

#Second quantization
new_qat_hparams = object_detector.QATHParams(learning_rate=0.9, batch_size=4, epochs=15, decay_steps=5, decay_rate=0.96)
model.restore_float_ckpt()
model.quantization_aware_training(train_data, validation_data, qat_hparams=new_qat_hparams)
#Skip validation because takes a long time
#qat_loss, qat_coco_metrics = model.evaluate(validation_data)
#print(f"QAT validation loss: {qat_loss}")
#print(f"QAT validation coco metrics: {qat_coco_metrics}")
```
